```python
#import all major libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data from CSV file
dataFrame= pd.read_csv('./Seattle_Real_Time_Fire_911_Calls.csv')
```

```python
#Check head of our Data Frame
dataFrame
```

```python
#top 10 reassons for reachig 911
dataFrame['Type'].value_counts().head(10)
```

```python
dataFrame['Type'].nunique()
```

```python
#Specific reasons to call 911
dataFrame['SpecificReason']=dataFrame['Type'].apply(lambda Type: Type.split(' ')[0] )
```

```python
dataFrame
```

```python
dataFrame['SpecificReason'].value_counts().head(10)
```

```python
#Plot a countplot for specific Reason
sns.countplot(x='SpecificReason',data= dataFrame.head(10))
```

```python
#Convert Datetime from object to actual date time object
dataFrame['Datetime']= pd.to_datetime(dataFrame['Datetime'])
```

```python
dataFrame.head()
```

```python
dataFrame.info()
```

```python
type(dataFrame['Datetime'].iloc[0])
```

```python
dataFrame['Hour']= dataFrame['Datetime'].apply(lambda time: time.hour)
dataFrame['Month']= dataFrame['Datetime'].apply(lambda time: time.month)
dataFrame['Day']= dataFrame['Datetime'].apply(lambda time: time.dayofweek)
```
